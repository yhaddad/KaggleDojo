```python
import matplotlib.pyplot as plt

%matplotlib inline

def diff_series(serie, interval=3):
    """Differenciate a time series."""
    diff = list()
    for i in range(interval, len(serie)):
        try:
            value = serie[i] - serie[i - interval]
            diff.append(value)
        except:
            continue
    return pd.Series(diff)
def invert_diff_series(history, yhat, interval=1):
    """Revert the differenciation."""
    return yhat + history[-interval]

def plot_serie(series, forecast=None, attr=""):
    """Debug plots about the focasting."""
    plt.figure(figsize=(16, 3))
    plt.plot(series, c='#38ABA5')
    if forecast is not None : 
        plt.plot(forecast, c='red')
    plt.tight_layout()
    plt.show()
```

```python
import statsmodels.api as sm
from sklearn import metrics
```

```python
steps_in_hour = 1
stationary_ = diff_series(serie, steps_in_hour)
print '[stationarity test] : {0}'.format(
        sm.tsa.stattools.adfuller(stationary_)
    )
```

```python
history = [ x for x in serie.values ]
predictions = []
for i in range(len(history)):
    # difference data
    steps_in_hour = 12
    diff = diff_series(serie, steps_in_hour)
    # predict
    model     = sm.tsa.ARIMA(diff.values, order=(4,0,4))
    model_fit = model.fit(trend='nc', disp=0)
    yhat      = model_fit.forecast()[0]
    yhat      = invert_diff_series(history, yhat, steps_in_hour)
    predictions.append(yhat)
    obs = serie.values[i]
    history.append(obs)
```

```python
plt.figure(figsize=(12,3))
plt.plot(history, label='test')
plt.plot(np.abs(predictions), label='prediction')

plt.legend()
```
