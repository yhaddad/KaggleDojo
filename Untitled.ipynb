```python
import numpy as np
```

```python
mu_s, sigma_s = 0.45,0.4  # mean and standard deviation
s_raw = np.random.normal(mu_s, sigma_s, 4000)

mu_b, sigma_b = -0.1,0.5  # mean and standard deviation
b_raw =np.random.normal(mu_b, sigma_b, 4000)

weight =np.concatenate((np.random.binomial(10,0.5, 4000)/100.0,np.random.binomial(10,0.5, 4000)/5.0))
y =np.concatenate((np.ones(s_raw.shape[0]),np.zeros(b_raw.shape[0])))

data = np.concatenate((s_raw,b_raw))
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig = plt.figure(figsize=(5,5))
plt.hist(data[y==0], bins=200, range=[-3,3],weights=weight[y==0], histtype='stepfilled', alpha=0.5)
plt.hist(data[y==1], bins=200, range=[-3,3],weights=weight[y==1], histtype='stepfilled', alpha=0.5)
plt.yscale('log')
plt.show()
```

```python
from scipy import optimize

class binner_base(object):
    """ Abstract class for classification based binning
    """
    def __init__(self, nbins, range):
        self.nbins = nbins
        self.range = range
        raise NotImplementedError('Method or function has not been implemented yet')
    def fit(self, X, y, sample_weights=None):
        raise NotImplementedError('Method or function has not been implemented yet')
        return self



class zbinner(binner_base):
    def __init__(self,nbins, range, drop_last_bin=True):
        self.nbins = nbins
        self.range = range
        self.drop_last_bin = drop_last_bin
        self.X = None
        self.y = None
        self.sample_weights = None
        
    def _cost_(self,s,b):
        c = np.true_divide( s, np.sqrt(b) )
        c[~np.isfinite( c )] = 0
        c[~np.isnull  ( s )] = 0
        return c
    def binned_score(self,x):
        nb_,_  = np.histogram(self.X[self.y==0],bins=x, range=self.range, weights=self.sample_weights[self.y==0])
        ns_,_  = np.histogram(self.X[self.y==1],bins=x, range=self.range, weights=self.sample_weights[self.y==1])
        return self._cost_(ns_, nb_)
    def minimize(self,x):
        x_ = np.sort(x)
        z  = self.binned_score(x_)
        return (-z**2).sum()
    def fit(self, X, y, sample_weights=None):
        self.X = X
        self.y = y 
        if sample_weights is not None:
            self.sample_weights = sample_weights
        else:
            self.sample_weights = np.ones(X.shape[0])
        bins_0 = np.linspace (self.range[0],self.range[1],self.nbins+1)
        print "[initial] bins         = ", bins_0
        print "[initial] significance = ", self.binned_score(bins_0)
        print "[initial] global sig   = ", self.minimize(bins_0)
        np.random.seed(555)   # Seeded to allow replication.
        min_args = {"method": "BFGS"}
        res = optimize.basinhopping(self.minimize, bins_0, minimizer_kwargs=min_args, niter=500)
        return res
```

```python
binner = zbinner(2,[-2,2])
results = binner.fit(data, y, sample_weights = weight)
```
